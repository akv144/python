```
data=""
for i in rdd.collect():
    data=data+i
data=data.split("|")
data
```

```
col=5
row=len(data)//5
row,col
```

```
data1=[]
ii=0
for i in range(0,row):
    r=[]
    for j in range(0,col):
        r.append(data[ii])
        ii=ii+1
    data1.append(r)
data1

        
```

```
df=spark.createDataFrame(data1,["name","EdU","YearOfExp","Tech","MobNo."])
```
