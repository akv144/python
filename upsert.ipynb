```
data = [[1,'Jamess','Smith'],[2,'Akv','verma'],[3,'Ankisha','Shindhe']]
sdf=spark.createDataFrame(data,["id","fname","lname"])
display(sdf)
```

```
spark.sql("create table if not exists upsert(id int,fname varchar(20),lname varchar(20)) using delta ")
```

```
%sql
select * from upsert
```

```
sdf.createOrReplaceTempView("source_view")
```

```
spark.sql("merge into upsert as target using source_view as source on target.id=source.id when matched then update set target.fname=source.fname,target.lname=source.lname when not matched then insert (id,fname,lname) values(id,fname,lname)")
```
